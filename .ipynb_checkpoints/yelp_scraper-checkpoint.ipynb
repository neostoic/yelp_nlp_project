## Solo Scrape Function

```python
url=['http://www.yelp.com/biz/jean-georges-new-york?start=%s']
page_order = range(0,401,40)

jg_reviews = []

for i in page_order:
    page = requests.get('http://www.yelp.com/biz/jean-georges-new-york?start=%s' % i)
    tree = html.fromstring(page.text)
    
    # This will create a list of reviews:
    reviews = tree.xpath('//p[@itemprop="description"]/text()')
    
    # This will add the reviews to a main list of reviews
    jg_reviews = jg_reviews + reviews
    
```

```python
import string
import pandas as pd

unicode_less_reviews = []

for s in jg_reviews:
    a = filter(lambda x: x in string.printable, s)
    unicode_less_reviews.append(a)

yelp_reviews = { 'yelp_reviews': unicode_less_reviews }

frame = pd.DataFrame(yelp_reviews)
frame.tail()
```
